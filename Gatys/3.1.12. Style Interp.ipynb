#### Style Interpolation

```python
# Imports
from train import train_model
from models import *
```

```python
# Values set from the Gatys paper 
alpha, beta = 1, 0.0001
weights = {'conv1_1': 0.2, 'conv2_1': 0.2, 'conv3_1': 0.2, 'conv4_1': 0.2, 'conv5_1': 0.2}
content_path = "data/content_images/geisel.jpg"
```

```python
# Multiple style with interpolation
lamdas = [(0.2, 0.8), (0.8, 0.2)]
style_paths = ["data/style_images/mosaic.jpg", "data/style_images/candy.jpg"]
for lamda in lamdas:
    train_model(style_paths[0], content_path, alpha, beta, 1, 0, 3000, weights, optimiser = 'LBFGS',  multiple_style_flag = 1, style_path_2 = style_paths[1], interp_value = lamda)
```
