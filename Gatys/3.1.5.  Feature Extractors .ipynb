```python
# Values set from the Gatys paper 
alpha, beta = 1, 0.0001
weights = {'conv1_1': 0.2, 'conv2_1': 0.2, 'conv3_1': 0.2, 'conv4_1': 0.2, 'conv5_1': 0.2}
style_path = "data/style_images/mosaic.jpg"
content_path = "data/content_images/amber.jpg"
```

```python
options = ['vgg_16','vgg_19','resnet','alexnet']
for opt in options:
    if opt == 'resnet':
        weights = {'layer1': 0.2, 'layer2': 0.2, 'layer3': 0.2, 'layer4': 0.2}
    elif opt =='alexnet':
        weights = {'conv1': 0.2, 'conv2': 0.2, 'conv3': 0.2, 'conv4': 0.2, 'conv5': 0.2} 
    
    train_model(style_path, content_path, alpha, beta, 1, 0, 1000, weights, optimiser = 'LBFGS', model_name = opt)
```
